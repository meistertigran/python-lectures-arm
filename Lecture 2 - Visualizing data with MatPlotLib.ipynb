# Գրաֆիկները ստեղծում MatPlotLib-ով

**Matplotlib**-ը Python-ի գրադարան է, որը թույլ է տալիս ստեղծել գրաֆիկներ։ Այն լայնորեն օգտագործվում է տվյալների պատկերավորման համար։ Matplotlib-ը հիմնականում գործածվում է որպես 2D գրաֆիկական գործիք, բայց ունի նաև 3D պատկերներ ստեղծելու հնարավորություններ:

Այն կարելի է են ներմուծել `plt` անվանմամբ հետևյալ հրահանգով։

````python
import matplotlib.pyplot as plt
````

MatPlotLib-ը ստեղծված է պատկերելու NumPy  զանգվածներ, այսինքն պետք է նաև ներմուծել Numpy-ն 

````python
import numpy as np
````

MatPlotLib-ով գրաֆիկներ ստեղծելու համար պետք է նրան տրամադրել պատկերվող տվյալները, թվային տեսքով։ Մեկ փոփոխականի Ֆունկցիային կորի տվյալները այդ ֆունկցիայի արժեքն է տարբեր կետերում։ Օրինակ ստեղծենք պարաբոլական ֆունկցիայի տվյալները.
$$
y = x^2
$$

````python
x = np.linspace(-10, 10, 21)
y = x**2

for i in range(1, 10):
    print('y(', x[i], ') = ', y[i])
````

Այս տվյալները որպես գրաֆիկի `x` և `y` առանցքների արժեքներ օգտագործելու համար դրան տրվում են `plt.plot()` ֆունկցիային որպես արգումենտ։ Այլնուհետև կարող ենք գրաֆիկը պատկերել `plt.show()` ֆունկցիայի միջոցով:

```python
plt.plot(x, y)
plt.show()
```

````python
plt.plot(x, y)
plt.show()
````

Ինպես տեսնում ենք, չնայած այն բանի, որ ֆունկցիային տրվել են դիսկրետ կետեր, այն պատկերում է հոծ գիծ։ Այս պահվածքը կարելի փոխել, սահմանելով կորի տեսքը որպես `plt.plot()` ֆունկցիայի երրորդ արգումենտ։ 

<pre style=""><code class="language-python">plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> y<span class="ͼ19">,</span> <mark class="add"><span class="ͼ12">'.'</span></mark><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">show</span><span class="ͼ19">(</span><span class="ͼ19">)</span></code></pre>

````python
def latex_arrow_props(color):
    return dict(
        arrowstyle='->,head_length=0.7,head_width=0.2',
        color=color,
        linewidth=1,
        shrinkA=2,
        shrinkB=8,
        connectionstyle="arc3,rad=0",
    )
````

````python
plt.plot(x, y, '.')
plt.annotate('Կետավոր գիծ', (x[6], y[6]), (x[6]+2, y[6]+20), color="green", arrowprops=latex_arrow_props("green"))
plt.show()
````

Առանցքների անվանումները տրվում են `plt.xlabel()` և `plt.ylabel()` ֆունկցիաներին։

<pre style=""><code class="language-python">plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> y<span class="ͼ19">)</span>

Ինչպես տեսնում ենք, նոր կորերը ավտոմատ կերպով պատկերվում են տարբեր գույնով, առանց մեր կողմից դա սահմանելու, սակայն կորի գույնը նույնպես կարողենք փոխել։

<pre style=""><code class="language-python">z <span class="ͼ10">=</span> np<span class="ͼ19">.</span><span class="ͼz">sin</span><span class="ͼ19">(</span>x<span class="ͼ19">)</span><span class="ͼ10">*</span><span class="ͼu">50</span> <span class="ͼ10">+</span> <span class="ͼu">50</span>

````python
x = np.linspace(-10, 10, 201)
y = x**2
z = np.sin(x)*50 + 50
````

Նույն գրաֆիկում մեկից ավել կորեր պատկերելիս, հաճախ անհրաժեշտ է լինում տալ կոնկրետ կորի մասին ինֆորմացիա։ Այդ նպատակով օգտագործվում է `.plot()` ֆունկցիայի `label` արգումենտը: `label`-ը սահմանելուց հետո, պետք է կանչել `.legend()` ֆունկցիան՝ լեգենդը ցուցադրելու համար:

<pre style=""><code class="language-python">plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> y<span class="ͼ19">,</span> <span class="ͼ12">'b-'</span><span class="ͼ19">,</span> <mark class="add">label<span class="ͼ10">=</span><span class="ͼ12">'$x^2$'</span></mark><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> z<span class="ͼ19">,</span> <span class="ͼ12">'r--'</span><span class="ͼ19">,</span> <mark class="add">label<span class="ͼ10">=</span><span class="ͼ12">'sin(x)'</span><span class="ͼ19"></mark>)</span>
plt<span class="ͼ19">.</span><span class="ͼz">xlabel</span><span class="ͼ19">(</span><span class="ͼ12">'My X axis'</span><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">ylabel</span><span class="ͼ19">(</span><span class="ͼ12">'My Y axis'</span><span class="ͼ19">)</span>
<mark class="add">plt<span class="ͼ19">.</span><span class="ͼz">legend</span><span class="ͼ19">(</span><span class="ͼ19">)</span></mark>

````python
plt.plot(x, y, 'b-', label='$x^2$')
plt.plot(x, z, 'r--', label='sin(x)')
plt.xlabel('My X axis')
plt.ylabel('My Y axis')
plt.annotate('Պիտակ', (8, 15), (6.5, 30), color="green", arrowprops=latex_arrow_props("green"))
plt.legend()
plt.show()
````

Գրաֆիկի վրա տեքստային մեկնաբանություններ ավելացնելու համար `plt.annotate()` ֆունկցիան է օգտագործվում: Այն թույլ է տալիս նշել գրաֆիկի կոնկրետ կետեր և դրանց վերաբերյալ տեղեկատվություն տրամադրել: Այն ընդունում է հետևյալ հիմնական արգումենտները.

1. `text`: Մեկնաբանության տեքստը:
2. `xy`: Կորի վրա կետի կոորդինատները, որի վերաբերյալ ուզում ենք թողնել մեկնաբանություն:
3. `xytext`: Մեկնաբանության տեքստի կոորդինատները (եթե տարբերվում է `xy`-ից):
4. `arrowprops`: Սլաքի հատկությունները, եթե ցանկանում ենք սլաք ավելացնել:
5. `fontsize`: Տեքստի չափը:
6. `color`: Տեքստի գույնը:

<pre style=""><code class="language-python">plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> y<span class="ͼ19">,</span> <span class="ͼ12">'b-'</span><span class="ͼ19">,</span> label<span class="ͼ10">=</span><span class="ͼ12">'$x^2$'</span><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">plot</span><span class="ͼ19">(</span>x<span class="ͼ19">,</span> z<span class="ͼ19">,</span> <span class="ͼ12">'r--'</span><span class="ͼ19">,</span> label<span class="ͼ10">=</span><span class="ͼ12">'sin(x)'</span><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">xlabel</span><span class="ͼ19">(</span><span class="ͼ12">'My X axis'</span><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">ylabel</span><span class="ͼ19">(</span><span class="ͼ12">'My Y axis'</span><span class="ͼ19">)</span>
<mark class="add">plt<span class="ͼ19">.</span><span class="ͼz">annotate</span><span class="ͼ19">(</span><span class="ͼ12">'min($x^2$)'</span><span class="ͼ19">,</span> <span class="ͼ19">(</span><span class="ͼu">0</span><span class="ͼ19">,</span> <span class="ͼu">0</span><span class="ͼ19">)</span><span class="ͼ19">,</span> <span class="ͼ19">(</span><span class="ͼu">0</span><span class="ͼ19">,</span> <span class="ͼu">20</span><span class="ͼ19">)</span><span class="ͼ19">,</span> color<span class="ͼ10">=</span><span class="ͼ12">"blue"</span><span class="ͼ19">,</span> fontsize<span class="ͼ10">=</span><span class="ͼu">12</span><span class="ͼ19">,</span> arrowprops<span class="ͼ10">=</span><span class="ͼ14">{</span><span class="ͼ12">"arrowstyle"</span>: <span class="ͼ12">"-&gt;"</span><span class="ͼ14">}</span><span class="ͼ19">)</span></mark>
plt<span class="ͼ19">.</span><span class="ͼz">legend</span><span class="ͼ19">(</span><span class="ͼ19">)</span>
plt<span class="ͼ19">.</span><span class="ͼz">show</span><span class="ͼ19">(</span><span class="ͼ19">)</span></code></pre>

````python
plt.plot(x, y, 'b-', label='$x^2$')
plt.plot(x, z, 'r--', label='sin(x)')
plt.xlabel('My X axis')
plt.ylabel('My Y axis')
plt.annotate('min($x^2$)', (0, 0), (0, 20), color="blue", fontsize=12, arrowprops={"arrowstyle": "->"})
plt.legend()
plt.show()
````

## Երկու և ավելի գրաֆիկների համատեղում մեկ պատկերում

Նախ ծանոթանանք MatPlotLib-ում առանցքների (axes) և պատկերների (figures) գաղափարի հետ:

1. Պատկեր (Figure):
   - Պատկերը MatPlotLib-ում ամբողջական գրաֆիկական պատուհանն է:
   - Այն կարող է պարունակել մեկ կամ մի քանի առանցքներ (axes):
   - Ստեղծվում է `plt.figure()` ֆունկցիայով:

2. Առանցքներ (Axes):
   - Առանցքները պատկերի մեջ գտնվող առանձին գրաֆիկական տարածքն են:
   - Յուրաքանչյուր axis ունի իր x և y առանցքները:
   - Մեկ պատկերում կարող են լինել մեկ կամ մի քանի axis:

Մեկից ավել գրաֆիկներ ստեղծելու համար, պետք է մեկ պատկերում համատեղենք մի քանի առանցք։ Դա իրականացվում է ենթագծապատկերների (subplots) միջոցով.

3. Ենթագծապատկերներ (Subplots):
   - Subplots-ը թույլ են տալիս մեկ պատկերի մեջ ստեղծել մի քանի առանցքներ:
   - Դրանք օգտակար են մի քանի գրաֆիկներ համեմատելու կամ կապակցված տվյալներ ցուցադրելու համար:

#### 1. Ցրման գրաֆիկ (Scatter plot):

Օգտագործվում է երկու փոփոխականների միջև հարաբերությունը ցույց տալու համար: Յուրաքանչյուր կետ ներկայացնում է մեկ դիտարկում:

````python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
````

#### 2. Սյունակային գրաֆիկ (Bar plot):

Ցույց է տալիս կատեգորիկ տվյալները ուղղահայաց կամ հորիզոնական սյունակների միջոցով:

````python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

plt.bar(fruits, counts, label=bar_labels, color=bar_colors)

plt.ylabel('fruit supply')
plt.title('Fruit supply by kind and color')
plt.legend(title='Fruit color')

plt.show()
````

````python
# Define a function to generate Z values
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Generate X, Y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create a filled contour plot
plt.figure()
cp = plt.contourf(X, Y, Z, levels=10, cmap='viridis')  # Filled contours
plt.colorbar(cp)  # Add color bar

# Overlay contour lines on top of the filled contours
contour_lines = plt.contour(X, Y, Z, levels=10, colors='black')
plt.clabel(contour_lines, inline=True, fontsize=8)

plt.title('Filled Contour plot of sin(sqrt(x^2 + y^2))')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Display the plot
plt.show()
````
