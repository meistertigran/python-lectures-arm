```python
import matplotlib.pyplot as plt
import numpy as np
```

Պատկերացրեք, որ դուք անշարժ գույքի գործակալ եք և ցանկանում եք հասկանալ կապը բնակարանների մակերեսի և գնի միջև Երևան քաղաքում: Այս դեպքում ցրման գրաֆիկը կարող է լինել գերազանց գործիք:
Ահա, թե ինչպես կարելի է այն իրականացնել:

```python
# Գեներացնենք հիպոթետիկ տվյալներ (իրական կյանքում այս տվյալները կգան տվյալների բազայից)
np.random.seed(42)
surface = np.random.randint(30, 150, 100)  # 100 բնակարան, 30-ից 150 քմ
price = 200000 + (surface * 1000) + np.random.normal(0, 20000, 100)  # Գինը՝ հիմնված մակերեսի վրա, պլյուս պատահական տատանում

# Ստեղծենք սփռման գրաֆիկը
plt.figure(figsize=(10, 6))
plt.scatter(surface, price / 1000000, alpha=0.6)
plt.xlabel('Բնակարանի մակերես (քմ)')
plt.ylabel('Գին (միլիոն դրամ)')
plt.title('Բնակարանի մակերեսի և գնի հարաբերակցությունը Երևանում')

# Ավելացնենք միտման գիծ
z = np.polynomial.Polynomial.fit(surface, price / 1000000, 1)
plt.plot(surface, z(surface), "r-", alpha=0.8)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
