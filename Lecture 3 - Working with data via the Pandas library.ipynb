# Տվյալների մշակում և վերլուծում Pandas գրադարանի միջոցով

Pandas-ը Python-ի հզոր գրադարան է տվյալների վերլուծության և մշակման համար: Այն առաջարկում է տվյալների կառուցվածքային երկու հիմնական տեսակ՝ հաջորդականություն (Series) միաչափ տվյալների և աղյուսակ (DataFrame) երկչափ տվյալների հետ աշխատելու համար։ Այն թույլ է տալիս արդյունավետորեն աշխատել կարգավորված տվյալների հետ:

Pandas-ի հիմնական առավելություններն են՝

1. Տվյալների արագ ընթերցում և գրում տարբեր ֆորմատներից (CSV, Excel, SQL և այլն)
2. Տվյալների արդյունավետ մաքրում և նախապատրաստում
3. Հզոր գործիքներ տվյալների ֆիլտրման, խմբավորման և միավորման համար
4. Ժամանակային հաջորդականությունների  վերլուծության լայն հնարավորություններ
5. Ինտեգրում այլ գիտական հաշվարկների գրադարանների հետ, ինչպիսիք են NumPy-ն և Matplotlib-ը

Pandas-ը լայնորեն օգտագործվում է տվյալագիտության, մեքենայական ուսուցման և ֆինանսական վերլուծության ոլորտներում:

Pandas գրադարանը ներմուծում ենք հետևյալ հրամանով։

```python
import pandas as pd
```

## Աշխատանք աղուսակների հետ DataFrame-ների միջոցով

"arm-salaries.xlsx" Excel ֆայլը պարունակում է տվյալներ Հայաստանի միջին աշխատավարձերի մասին: Այս տվյալները ներկայացված են եռամսյակային կտրվածքով, սկսած 2018 թվականից: Տվյալները վերցվել են [armstat.am](https://www.armstat.am/) կայքից։

Նախքան սկսելը, անհրաժեշտ է կատարել հետևյալ նախապատրաստական քայլերը՝

1. Ստեղծեք նոր պանակ (folder) "data" անունով ձեր Jupyter notebook-ի գտնվելու պանակում:
2. Տեղադրեք "arm-salaries.xlsx" ֆայլը այդ նոր ստեղծված "data" պանակի մեջ:

Այս քայլերը կարևոր են ֆայլերի կազմակերպված կառուցվածք ունենալու և հետագայում տվյալները հեշտությամբ հասանելի դարձնելու համար: Այս կերպ, երբ դուք օգտագործեք `pd.read_excel()` ֆունկցիան ֆայլը բեռնելու համար, կարող եք հղում անել ֆայլին հետևյալ  ուղիով (path)՝ "data/arm-salaries.xlsx":

[pd.read_excel()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html) ֆունկցիան Pandas գրադարանի կարևոր գործիքներից մեկն է, որը թույլ է տալիս հեշտությամբ ներմուծել տվյալներ Excel ֆայլերից: Այս ֆունկցիան օգտագործելով՝ դուք կարող եք արագորեն և արդյունավետորեն բեռնել "arm-salaries.xlsx" ֆայլի բովանդակությունը DataFrame-ի մեջ:

```python
pd.read_excel("data/arm-salaries.xlsx")
```

Սակայն այս աղյուսակը պարունակում է բազմաթիվ տվյալներ, որոնք մեզ անհրաժեշտ չեն։ Որպես վերջնարդյունք ուզում ենք ունենալ `E`-ից `AD` սյունակներում և 7-ից 18 տողերում պարունակվող տվյալները։ Ուզում ենք նաև շրջել աղյուսակը այնպես, որ ամեն սյունակները պարունակեն ամեն մարզի աշխատավարձը, իսկ տողերը տվյալ եռամսյակի, օգտագործելով մարզերի անգլերեն անվանումները որպես սյուների անվանում, իսկ ամեն եռամսյակի ամսաթիվը որպես տողերի անուն։

Սկսենք մաքրել տվյալները։ Նախ ընտրենք մեզ անհրաժեշտ թերթը։ `sheet_name` արգումենտը թույլ է տալիս նշել, թե Excel ֆայլի որ թերթից (sheet) պետք է կարդալ տվյալները: Այնուհետև նշենք, որ տվյալ դեպքում Excel-ի ֆայլը չի պարունակում սյունակների անվանումներ, քանի-որ դրանք ինքներս ենք ավելացնելու հետագա քայլերում։ `header` արգումենտն է, որ որոշում է, թե որ տողն է պարունակում սյունակների անվանումները: Տալով նրան `None` հատուկ արժեքը, նշում ենք, որ սյուները չունեն անվանումներ։

```python
pd.read_excel("data/arm-salaries.xlsx", sheet_name=0, header=None)
```

Նշելու համար թե, որ սյունակներն են մեզ անհրաժեշտ աղյուսակից, օգտագործենք `usecols` և `index_col` արգումենտները։ `usecols` արգումենտը թույլ է տալիս նշել, թե կոնկրետ որ սյունակները պետք է ներմուծել Excel ֆայլից: 
`index_col` արգումենտը օգտագործվում է նշելու համար, թե որ սյունակը պետք է օգտագործվի որպես DataFrame-ի ինդեքս, այսինք, տողերի անվանում: Նշենք, որ `index_col`-ի դեպքում, պետք է նշել արդեն ֆլիտրված տվյալներում տվյալի տողի ունեցած համարը։

```python
pd.read_excel("data/arm-salaries.xlsx", sheet_name=0, header=None, usecols="C,E:AD",  index_col=0)
```

`skiprows` արգումենտի միջոցով կարող ենք ֆիլտեր աղյուսակի սկզբից n տող, իսկ `skipfooter` միջոցով աղյուսակի վերջից։

```python
pd.read_excel("data/arm-salaries.xlsx", sheet_name=0, header=None, usecols="C,E:AD", index_col=0, skiprows=7, skipfooter=6)
```

Սյունակների անվանումները ստեղծելու համար, օգտվենք [pd.date_range](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html) ֆունկցիայից։ Այս ֆունկցիան թույլ է տալիս գեներացնել սահմանված ժամանակահատվածի համար ամսաթվերի հաջորդականություն որոշակի հաճախությոամբ: Այն վերադարձնում է [DatetimeIndex](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.html) օբյեկտ, որը Pandas-ի հատուկ ինդեքսի տեսակ է, նախատեսված ամսաթվերի և ժամանակների հետ աշխատելու համար: Այս ֆունկցիային որպես արգումենտ տալիս ենք ժամանակահատվածի սկիզբը և վերջը, որպես `string` (տեքստ) և `freq` արգումենտի միջոցով սահմանում ենք վերադարձվող ամսաթվերի հաճախությունը։ Քանի-որ մեր տվյալները եռամսյակային են, օգտագործել ենք `'QS'` հաճախությունը, որը նշանակում է "Quarter Start" ("Եռամսյակի սկիզբ")։

```python
pd.date_range("20180101", "20240601", freq="QS")
```

Ստանանք վերջնական սյունակների անվանումները զանգվածի տեսքով։ Առաջին սյան անվանումը պետք է թողնել դատարկ, քանի-որ այն պարունակում է մարզերի անվանումները։ Մնացած սյուների անվանումները ամսաթվերն են։ Քանի-որ [DatetimeIndex](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.html)-ը նաև numpy-ի զանգված է, նրանից զանգված ստանալու համար կարող ենք կանչել `.tolist()` ֆունկցիան։

```python
quarters = [''] + pd.date_range("20180101", "20240601", freq="QS").tolist()
quarters[:5]
```

[timestamp](https://en.wikipedia.org/wiki/Unix_time)-ը կոնկրետ պահի ճշգրիտ գրանցումն է ժամանակի մեջ: Այն սովորաբար ներկայացվում է որպես թվային արժեք, որը ցույց է տալիս, թե քանի վայրկյան է անցել ժամանակի որոշակի սկզբնակետից (սովորաբար Unix Epoch-ից, որը 1970 թվականի հունվարի 1-ն է) մինչև տվյալ պահը: timestamp-ը օգտագործվում է համակարգչային համակարգերում իրադարձությունները ճշգրիտ ժամանակագրելու համար:

Աղյուսակի սյուների անվանումները տալու համար, պետք է ընդամենը մեր `pd.read_excel()` ֆունկցիայի `names` արգումենտին տանք մեր նոր գեներացված զանգվածը։

```python
df = pd.read_excel("data/arm-salaries.xlsx", sheet_name=0, header=None, usecols="C,E:AD", index_col=0, skiprows=7, skipfooter=6, names=quarters)
df
```

Մեր ցանկացած վերջնարդյունքը այն էր, որ սյուներում ունենայինք ամեն մարզի աշխատավարձը, այսինքն անհրաժեշտը նաև շրջել աղյուսակը։ Դա կարող ենք իրականացնել `transpose` ֆունկցիայի միջոցով։

```python
df = df.transpose()
df
```

Տպենք միայն մեկ սյան արժեքները։

```python
df['Yerevan city']
```

Սակայն Արագածոտնի մարզի դեպքում, ինչ-որ պատճառով ստանում ենք սխալ։

```python
df['Aragatsotn']
```

Սա պայմանավորված է այն հանգամանքով, որ Excel-ի ֆայլում մարզերի անվանումների սկզբում կամ վերջում առկա են բացատներ, որոնք մեզ տեսանելի չեն։

![spaces in column names](images/spaces-in-names.png)

```python
df['Aragatsotn            ']
```

Տեքստի սկզբում և վերջում եղած բացատները կարելի մաքրել `str.strip()` ֆունկցիայի միջոցով։

```python
test_str = '              text              '
test_str
```

```python
test_str = str.strip(test_str)
test_str
```

Սյունակների անվանումների [Index](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Index.html) զանգվածը հասանելի է `df.columns`-ի միջոցով։

```python
df.columns
```

Անհրաժեշտ է 

1. այս զանգվածի ամեն անդամի վրա կանչել `str.strip` ֆունկցիան,
2. ստանալ նոր զանգված մաքրված անվանումներով,
3. `df.columns`-ին վերագրել նոր զանգվածը։

Index զանգվածի վրա կանչելով `.map()` ֆունկցիան կարող ենք նրան որպես արգումենտ տալ կամայական այլ ֆունկցիա և կանչել այն զանգվածի ամեն անդամի վրա։

```python
df.columns = df.columns.map(str.strip)
df.columns
```

Այսպիսով ավարտեցինք տվյալների մշակման փուլը և կարողեն սկսել վերլուծել դրանք։

## Տվյալների վերլուծում

*Առկա տվյալներով հաշվենք հայաստանում միջին աշխատավարձը ամեն եռամսյակի համար։*

`df.mean()` ֆունկցիայի միջոցով կարող ենք հաշվել միջին արժեք, ըստ տողերի և սյունակների։ Ըստ սյուների միջին արժեք հաշվելու համար կարող ենք ֆունկցիան կանչել առանց արգումենտների։ Սա համապատասցանում է `axis` արգումենտի `0` արժեքին և թելադրում է ֆունկցիային, որ գործողությունը պետք է իրականացնել ըստ սյուների

```python
df.mean()
```

Ըստ տողերի միջին արժեք հաշվելու համար, պետք է `axis` արգումենտին տալ `0` արժեքը։ Սա կհամապատասխանի հայաստանում միջին աշխատավարձին տվյալ եռամսյակում։

```python
df.mean(axis=1)
```

Այս արդյունքի վրա կանչելով `.round()` ֆունկցիան կկլորացնենք արժեքները։

```python
df.mean(axis=1).round()
```

Որպեսզի այս արժեքը ավելացնենք, որպես նոր սյուն մեր աղյուսակում, պարզապես պետք է այն վերագրենք նոր սյանը, նույն կերպ ինչ վերագրում ենք նոր արժեք ՝ `df['Armenia'] = df.mean(axis=1).round()`։

```python
df['Armenia'] = df.mean(axis=1).round()
df
```

Հաճախ չենք ուզում տեսնել ամբողջ աղյուսակի տվյալները, այլ միայն դրա մի հատվածը։ 

`df.tail()` ֆունկցիայի միջոցով կարողենք տպել աղյուսակի վերջին մի քանի տողը։

```python
df.tail()
```

`df.head()` ֆունկցիայի միջոցով կարողենք տպել աղյուսակի առաջին մի քանի տողը։

```python
df.head()
```

*Դուրս բերենք ամեն եռամսյակի ամենաբարձր աշխատավարձով մարզը և համապատասխան աշխատավարձը։*

`df.max(axis=1)` կանչով, կստանանք առավելագույն արժեքը ամեն տողում, այսինք, տվյալ եռամսյակի առավելագույն աշխատավարձը։ Հիշենք, որ հենց `axis=1` արգումնետի շնորհիվ է, որ գործողությունը իրականացվում է ըստ տողերի։ Դրա բացակայության դեպքում, այն կիրականցվեր ըստ սյուների։

```python
quarter_max_salary = df.max(axis=1).to_frame('Salary')
quarter_max_salary.tail()
```

[.to_frame()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.to_frame.html) ֆունկցիան օգտագործում ենք հաջորդականության օբյեկտը ([Series](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html)) աղյուսակի օբյեկտի վերածելու համար ([DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)) դրա հետ ավելի հեշտ աշխատելու համար։ Այն ընդունում է մեկ արգումենտը, որը նոր աղյուսակում սյան անվանումն է։

Սակայն մեզ նաև անհրաժեշտ է իմանալ, թե որ մարզում էր դիտվել առավելագույն աշխատավարձը։ Դա իրականացնելու համար, ընդամենը անհրաժեշտ է օգտվել `df.idxmax()` ֆունկցիայից, որը վերադարձնում է առավելագույն արժեքի տողի կամ սյան անվանումը, կախված `axis` արգումենտի արժեքից։

```python
quarter_max_salary_region = df.idxmax(axis=1).to_frame('Region')
quarter_max_salary_region.tail()
```

Վերջապես այս երկու աղյուսակները կարող ենք միացնել, օգտագործելով `.join()` ֆունկցիան։ Եթե կանչենք `df1.join(df2)` ապա `df1` աղյուսակին կկցվի `df2` աղյուսակը և կվերադարձվի նոր աղյուսակ, առանց փոփոխելու `df1`-ը։

```python
max_quarter = quarter_max_salary_region.join(quarter_max_salary)
max_quarter.tail()
```

Ստացված արդյունքը կարող ենք պահպանել Excel ֆորմատով `.to_excel()` ֆունկցիայի միջոցով, սակայն այդ դեպքում եռամսյակներին համապատասխանող ամսաթվերը դժվար ընթեռնելի են։

```python
max_quarter.to_excel('output/arm-salaries-max-bad.xlsx')
```

Իրավիճակը կարելի է լավացնել, օգտվելով նույն `.map()` ֆունկցիայից, որով մաքրել էինք մարզերի անվանումներում բացատները, սակայն այս դեպքում, ամսաթիվը կփոխարինենք "տարեթիվ եռամսյակ" տեսքով։

- `//` գործողությունը վերադարձնում է բաժանման գործողության ամբողջ մասը։ `7 // 3 = 2`
- `f"{var} text"`. `f` հատուկ օպերատորը թույլ է տալիս հեշտությամբ միավորել տեքստը և փոփողականները։

```python
def to_quarterly(date):
    quarter = (date.month - 1) // 3 + 1
    return f"{date.year} Q{quarter}"

max_quarter.index = max_quarter.index.map(to_quarterly)
max_quarter.tail()
```

Սա կարող ենք իրականացնել [pd.MultiIndex.from_tuples()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.MultiIndex.from_tuples.html) ֆունկցիայի միջոցով։ Այն որպես արգումենտ ընդունում է զույգերի (`tuple`) զանգված։ Նախ առանձնացնենք եռամսյակները տարեթվերից։ `.split()` ֆունկցիան տեքստը վերածում է զանգվածի բաժանելով այն բոլոր այն կետերում որտեղ կան զանգավածներ։

```python
max_quarter.index.str.split().to_list()
```

Այնուհետև վերածենք տարթիվը և եռամսյակը պարունակող զանգավածները tuple-ների։ `tuple()` ֆունկցիան, զանգվածը վերածում է tuple-ի։ Այն կարող ենք ազդել զանգվածի ամեն անամի վրա օգտվելով `for`-ի կարճ գրելաձևից։ 
`[func(a) for a in arr]`


```python
[tuple(a) for a in max_quarter.index.str.split().to_list()]
```

Այս արդյունքը տալով որպես արգումենտ `pd.MultiIndex.from_tuples()` ֆունկցիային և վերգրելով վերադարձվող արժեքը որպես նոր տողերի անվանում `max_quarter.index`-ին, կստանանք ցանկալի արդյունքը։

```python
max_quarter.index = pd.MultiIndex.from_tuples([tuple(a) for a in max_quarter.index.str.split().to_list()])
max_quarter.tail()
```

Մեզ հայտնի մեթոդներով այժմ ստեղծենք աղյուսակ որը պարունակում է բոլոր մարզերի համար միջին, առավելագույն և նվազագույն նշված ժամանակահատվածի համար։ Դասավորենք դրանք ըստ միջին աշխատավարձի աճման կարգով։

```python
regions_analysis.to_excel('output/arm-salaries-regions.xlsx', sheet_name='Region Analysis')
```
